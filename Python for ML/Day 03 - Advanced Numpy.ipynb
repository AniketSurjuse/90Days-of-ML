### Broadcasting

The term broadcast describe how Numpy treats arrays with different shapes during arithmatic operation. The smaller array is "broadcast" across the larger array so that they have compatible shapes

```python
a = np.arange(12).reshape(3,4)
b = np.arange(12,24).reshape(3,4)
```

```python
a+b
```

```python
b = np.arange(4)

#(3,4) and (4,)---> (1,4)---> (3,4)---> same shape hence operation takes place
```

```python
a+b
```

### Working with mathematical formulae

```python
#Sigmoid

def sigmoid(a):
    return 1/(1+np.exp(-a))

a = np.arange(10)
sigmoid(a)
```

```python
# Mean Square error

def mse(pred,actual):
    return np.mean((pred-actual)**2)

actual = np.random.randint(1,50,25)
predicted = np.random.randint(1,50,25)

mse(predicted,actual)
```

```python
#categorical cross entropy

def cce(pred,act):
    return np.mean((act*np.log(pred)-(1-act)*np.log(1-pred)))
```

```python
cce(predicted,actual)
```
